```python
# Define a function to process each file
def process_dataset(file_path, year_month):
    try:
        # Load the dataset with Latin-1 encoding
        df = pd.read_csv(file_path, encoding='latin1')
        # Add 'Year_Month' column
        df['Year_Month'] = year_month
        # Filter for 'Cuautla' municipality
        df_cuautla = df[df['municipio'].str.lower() == 'cuautla']
        return df_cuautla
    except Exception as e:
        return e

# File paths and corresponding year_month values
file_paths = [
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_0419.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_0420.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_0521.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_0522.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_1118.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_1119.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_1120.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_17_1121.csv',
    '/Users/jairmartinez/Documents/UPY/Data mining/Final Project Data Mining/Data Frames/denue_inegi_1122.csv'
]

year_months = [
    '04_2019',
    '04_2020',
    '05_2021',
    '05_2022',
    '11_2018',
    '11_2019',
    '11_2020',
    '11_2021',
    '11_2022'
]

# Initialize a list to hold dataframes
dfs = []

# Process each dataset and collect them in the list
for path, ym in zip(file_paths, year_months):
    df_processed = process_dataset(path, ym)
    # Only append if the returned value is a DataFrame
    if isinstance(df_processed, pd.DataFrame):
        dfs.append(df_processed)

# Checking how many datasets were processed successfully
len(dfs)
```

## Check unique values in nombre_act and nom_estab to see the new bussinesses 

```python
# Group the data by 'nombre_act' and 'Year_Month' and count unique 'nom_estab'
business_count_by_date = df.groupby(['nombre_act', 'Year_Month'])['nom_estab'].nunique().reset_index()

# Rename the 'nom_estab' column to 'business_count'
business_count_by_date.rename(columns={'nom_estab': 'business_count'}, inplace=True)

# Pivot the data to have 'Year_Month' as columns and 'nombre_act' as rows
business_evolution = business_count_by_date.pivot(index='nombre_act', columns='Year_Month', values='business_count').fillna(0)

# Display the first few rows of the pivoted dataframe
business_evolution.head(20)
```

### Visualizing the growth of this bussinesses by importance and amount of growth or fall

```python
import matplotlib.pyplot as plt

# Calculate the difference in business count from the first to the last period for each activity
business_evolution['change'] = business_evolution.iloc[:, -1] - business_evolution.iloc[:, 0]

# Sort activities by the absolute value of their change to get the ones with the most significant changes
top_activities = business_evolution['change'].abs().sort_values(ascending=False).head(10).index

# Filter the dataframe to include only the top activities
top_activities_evolution = business_evolution.loc[top_activities]

# Drop the 'change' column as it was only needed for sorting
top_activities_evolution = top_activities_evolution.drop(columns='change')

# Plotting
plt.figure(figsize=(14, 8))

for activity in top_activities:
    plt.plot(top_activities_evolution.columns, top_activities_evolution.loc[activity], marker='o', label=activity)

plt.title('Evolution of Business Count for Top Activities')
plt.xlabel('Year_Month')
plt.ylabel('Number of Businesses')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### Dates are not sorted so I have to sort them out

```python
# Convert the 'Year_Month' columns to datetime to sort them
top_activities_evolution.columns = pd.to_datetime(top_activities_evolution.columns, format='%m_%Y')

# Sort the columns by date
top_activities_evolution = top_activities_evolution.reindex(sorted(top_activities_evolution.columns), axis=1)

# Plotting
plt.figure(figsize=(20, 12))

for activity in top_activities:
    plt.plot(top_activities_evolution.columns.strftime('%m_%Y'), top_activities_evolution.loc[activity], marker='o', label=activity)

plt.title('Evolution of Business Count for Top Activities')
plt.xlabel('Year_Month')
plt.ylabel('Number of Businesses')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
